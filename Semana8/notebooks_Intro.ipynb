```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rc
from mpl_toolkits.mplot3d import axes3d
from matplotlib import cm
import matplotlib as mpl
```

```python
h = 0.02
v_0 = 1
v= 0.2
w = 0.5

R = v_0 * h/v

Min_x, Max_x, N = 25.,30.,51
x = np.linspace(Min_x,Max_x,N)
Min_y, Max_y = 5.,5.
y = np.linspace(Min_y,Max_y,N)

u = np.zeros((N,N))
omega = np.zeros((N,N))

M,N = u.shape
for i in range(M):
    for j in range(N):
            
        if j == 0:
            u[i,j] = 0
            omega[i,j] = -2*(u[i,j+1] - u[i,j])
        elif j == N - 1:
            u[i,j] = 0
            omega[i,j] = -2*(u[i,j-1] - u[i,j])
        else:
            u[i,j] += (u[i,j+1] + u[i,j-1])/4.
            omega[i,j] += (omega[i,j+1]+omega[i,j-1])/4 - (R/16)*(u[i,j+1]*u[i,j-1])    
                
        if i == 0:
            omega[i,j] = -2*(u[i+1,j] - u[i,j])
   
        elif i == M - 1:
            omega[i,j] = -2*(u[i-1,j] - u[i,j])    
   
        else:
            omega[i,j] += (omega[i+1,j] + omega[i-1,j] )/4 + (R/16)*(u[i+1,j]*u[i-1,j])
            u[i,j] += (u[i+1,j] + u[i-1,j])/4
            
z = np.zeros_like(u)
fig = plt.figure()
ax = fig.add_subplot(1,3,1)
ax1 = fig.add_subplot(1,3,2)
ax2 = fig.add_subplot(1,3,3)
ax.set_title('1')
ax.streamplot(x,y,u,w,color='b',density=2)
ax1.set_title('2')
ax1.contour([u,w],z)
```
